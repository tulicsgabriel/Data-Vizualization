```python
# Standard import
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# set style
sns.set_style("whitegrid") # style must be one of white, dark, whitegrid, darkgrid, ticks
```

```python
# basic distribution plot
dataset = randn(100)
sns.displot(dataset, bins = 25)
```

```python
# While in histogram mode, it is also possible to add a KDE curve, set the kde to True 
sns.displot(dataset, bins = 25, kde=True, color = 'tab:red')
```

```python
# we can also set the kind parameter, kind` must be one of ['hist', 'kde', 'ecdf']
# ecdf ->  empirical cumulative distribution functions

sns.displot(dataset, kind="ecdf", rug = True)
```

```python
# Seaborn work perfectly with Series and DataFrames as well, here is an example
ser = pd.Series(dataset, name = 'Random numbers')
ser
```

```python
# Adjust font family and font size
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 14})

sns.displot(ser, bins = 25, kde = True, color = 'tab:purple', label="Stuff")
# show legend
plt.legend()
plt.show()
```
