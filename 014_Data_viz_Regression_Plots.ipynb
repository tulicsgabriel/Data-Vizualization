```python
# Standard import
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# set style
sns.set_style("darkgrid") # style must be one of white, dark, whitegrid, darkgrid, ticks
```

```python
# use a built-in dataset called 'tips'
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# plot 
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# Adjust font family and font size
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 14})

# Adjust marker and linewidth
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,
           scatter_kws = {'marker':'o', 'color':'indianred'}, 
           line_kws = {'linewidth': 1, 'color': 'blue'}
          )
```

```python
# change order of polynomial regression, in case if you don't want a linear fit

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, order = 4,
           scatter_kws = {'marker':'o', 'color':'indianred'}, 
           line_kws = {'linewidth': 1, 'color': 'blue'}
          )
```

```python
# Scatter plot between the two variables
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
```

```python
# same thing with the scatterplot command
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
# change the style of the markers
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="time")
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', palette="Set2")
```

```python
# combining plots in two subplots
fig, (axis1, axis2) = plt.subplots(1, 2, sharey = True)

sns.regplot(x = 'total_bill', y = 'tip', data = tips, ax = axis1)
sns.stripplot(x = 'sex', y = 'tip', data = tips, hue = 'sex', palette="Set2", ax = axis2)
```
