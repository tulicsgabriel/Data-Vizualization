## Array Processing

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.arange(-5, 5, 0.01) # start, stop, step
```

```python
dx, dy = np.meshgrid(points, points) # Shift + Tab for more information
# Return coordinate matrices from coordinate vectors.
# https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html
dx
```

```python
# column-wise -> dy, this will help us to make our grid
dy
```

```python
z = (np.sin(dx) + np.sin(dy)) # evaluation function
z
```

```python
# we can plot this
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('Plot of sin(x) + sin(y)')
```

### Numpy where

```python
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
condition = np.array([True, True, False, False])
```

```python
# list comprehension: we choose value from A if condition is True, and B otherwise
# zip is for multiple array iterations
answer = [(A_val if cond else B_val) for A_val, B_val, cond in zip(A, B, condition)]
answer
```

```python
# short version of the above
answer2 = np.where(condition, A, B)
answer2
```

```python
# another example
from numpy.random import randn
arr = randn(5,5)
arr
```

```python
np.where(arr<0, 0, arr)
```

```python
arr = np.arange(1,10)
arr.reshape(3,3)
print(arr.shape)
print(arr)
# this doesn't reshape the actual array
```

```python
# but this does, be carefull
arr = arr.reshape(3,3)
arr
```

```python
# sum up every value
print(arr.sum())
```

```python
# sum up on specified axis
arr.sum(0)
```

```python
arr.sum(1)
```

```python
# mean
arr.mean()
```

```python
arr.mean(1)
```

```python
# standard deviation
arr.std()
```

```python
# variance
arr.var()
```

## any, all

```python
bool_arr = np.array([True, False, True, True, False, True])

# any -> returns True if ANY element in the array is True
bool_arr.any()
```

```python
# all -> any -> returns True if ALL element in the array is True
bool_arr.all()
```
