```python
# Standard import
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# set style
sns.set_style("dark") # style must be one of white, dark, whitegrid, darkgrid, ticks
```

```python
dataset = randn(100)


plt.hist(dataset, density = True, color = 'tab:red', alpha = 0.5, bins = 10)
sns.kdeplot(dataset)
```

```python
# KDE for multidimensional data

# center of the data
mean = [0, 0]

# covariance
cov = [[1, 0], [0, 100]]

# random samples from a multivariate normal distribution
dataset2 = np.random.multivariate_normal(mean, cov, 1000)
#indexes = np.arange(len(dataset2))

dframe = pd.DataFrame(dataset2, columns = ['X', 'Y'])

dframe.head()
```

```python
sns.kdeplot(x = dframe.X, y = dframe.Y, shade = True)
```

```python
# different bandwidht method
sns.kdeplot(x = dframe.X, y = dframe.Y, shade = True, bw_method = 10)
```

```python
# kde joint plot
sns.jointplot(x = dframe.X, y = dframe.Y, kind= 'kde')
```
