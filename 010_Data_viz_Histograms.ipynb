```python
# Standard import
import numpy as np
import pandas as pd
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# set style
sns.set_style("darkgrid") # style must be one of white, dark, whitegrid, darkgrid, ticks
```

```python
dataset1 = randn(100)
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color = 'indianred')
```

```python
# both histograms in one plot, with labels 
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Histogram of random numbers')

# Tableau  blue and red, becauese it's fancy
plt.hist(dataset1, density = True, color = 'tab:red', alpha = 0.5, bins = 20)
plt.hist(dataset2, density = True, color = 'tab:blue', alpha = 0.5, bins = 20)
```

```python
# Jointplot with seaborn

data1 = randn(1000)
data2 = randn(1000)


sns.jointplot(x = data1, y = data2, kind="resid") # kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
```

```python
# I want labels on the x and y axis, this is a work around
# + a little extra customazation

data1 = randn(1000)
data2 = randn(1000)

# Adjust font family and font size
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 16})

jointplotobject = sns.jointplot(x = data1, y = data2, kind="hist", color = 'violet')

# JointGrid has a convenience function
jointplotobject.set_axis_labels('Random data 1', 'Random data 2')

```
