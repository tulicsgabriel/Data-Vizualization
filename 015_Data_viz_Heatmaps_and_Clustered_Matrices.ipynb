```python
# pivot the table
flight_frame = flight_frame.pivot('month', 'year', 'passengers')
flight_frame
```

```python
# create a heatmap of the data
sns.heatmap(flight_frame)
```

```python
yearly_flights = flight_frame.sum()
years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_frame = pd.concat((years, flights), axis = 1)
year_frame.columns = ['Years', 'Flights']

year_frame
```

```python
# combining plots in two subplots: barplot and a heatmap
fig, (axis1, axis2) = plt.subplots(2, 1)

sns.barplot(x = 'Years', y = 'Flights', data = year_frame, ax = axis1)
sns.heatmap(flight_frame, cmap = 'Blues', ax = axis2)
```

```python
# combining plots in two subplots: barplot and a heatmap
fig, (axis1, axis2) = plt.subplots(2, 1)

sns.barplot(x = 'Years', y = 'Flights', data = year_frame, ax = axis1)
sns.heatmap(flight_frame, cmap = 'Blues', ax = axis2, cbar_kws = {'orientation':'horizontal'})
```

## Clustermaps  

```python
sns.clustermap(flight_frame, cmap = 'Blues')
```

```python
# only cluster by the months
sns.clustermap(flight_frame, col_cluster = False, cmap = 'Blues')
```

```python
# only cluster by the years
sns.clustermap(flight_frame, row_cluster = False, cmap = 'Blues')
```

```python
# since the number of number of flights increase every year we should set a standard scale
# standard_scale = 1 -> standardazi by the column / by the year

sns.clustermap(flight_frame, standard_scale = 1, cmap = 'Blues')

# now we can tell that the summer months are the most 'heavy' in fligh numbers
```

```python
# standard_scale = 0 -> standardazi by the row / by the months

sns.clustermap(flight_frame, standard_scale = 0, cmap = 'Blues')
```
